```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

```python
# Exercício 4.11

plt.figure(figsize=(15, 5))

f = np.zeros(256)
for i in range(128):
  f[i] = m.sin(10*m.pi*i/256)
for i in range(128):
  f[128+i] = 0.5

M = [256, 100, 50, 20, 10, 5, 1]
for m in M:
  v = np.zeros(256)
  for i in range(m):
    v[i] = 1/m
  for i in range(256-m):
    v[m+i] = 0

  x = np.real(np.fft.ifft(np.fft.fft(v)*np.fft.fft(f)))
  plt.plot(x, label='M = '+str(m))

plt.legend()
plt.show()
```

```python
# Exercício 4.13

h = np.array([1, -1, 1, 2])
H = np.fft.fft(h)
H = np.sort(H)
Mh = scipy.linalg.circulant(h)

autovalores, autovetores = np.linalg.eig(Mh)
autovalores = np.sort(autovalores)

# print(H, autovalores, autovetores, sep='\n\n')

assert np.allclose(autovalores, H)
print('H e autovalores de Mh são iguais!')
```
